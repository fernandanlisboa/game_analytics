```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### Metacritic

```python
#### Metacritic data

meta_df = pd.read_csv('meta_full_third.csv', sep=',', encoding='utf-8', header=0)
meta_df.drop(columns=['Unnamed: 0'], inplace=True)
meta_df
```

```python
meta_df = meta_df[['meta_game_name', 'meta_platform', 'meta_release_date', 'meta_critic_score', 'meta_user_score', 'meta_esrb', 'meta_genre', 'meta_multiplayer']]
meta_df
```

```python
meta_action = meta_df.loc[(meta_df['meta_genre'] == 'Action') | (meta_df['meta_genre'] == 'Action Adventure')]
meta_action
```

```python
columns = meta_action.columns
```

```python
meta_action['meta_game_name'].dtype
```

```python
for c in columns:
    if meta_action[c].dtype == 'O': 
#         meta_action.loc[meta_action[c].str.contains(pat='\n') == True, [c]] = meta_action[c].str[1:-1]
        meta_action.loc[meta_action[c].str.contains(pat='\n') == True, [c]] = meta_action[c].str.replace('  ', '')
        meta_action.loc[meta_action[c].str.contains(pat='\n') == True, [c]] = meta_action[c].str.replace('\n', '')
    
meta_action
```

```python
#  separando datas

meta_action['release_month'] = meta_action['meta_release_date'].str[:3]

meta_action['release_day'] = meta_action['meta_release_date'].str[3:6]

meta_action['release_year'] = meta_action['meta_release_date'].str[-4:]

meta_action['release_month'].replace({'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}, inplace=True)

meta_action['meta_multiplayer'].replace({'yes': 1, 'no': 0}, inplace=True)

meta_action
```

```python
meta_action.release_month.unique()
```

```python
meta_action.drop(columns=['meta_release_date', 'meta_esrb'], inplace=True)
```

```python
meta_action.to_csv('meta_games_cleaned.csv', sep=';', index=False, encoding='utf-8')
```

```python
meta_action.plot.scatter(y='meta_critic_score', x='meta_game_name')
```
